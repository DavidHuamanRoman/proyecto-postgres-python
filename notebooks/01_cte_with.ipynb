```python
import sys
import os
sys.path.append(os.path.abspath(os.path.join(os.getcwd(), os.pardir)))
import pandas as pd
from conexion import get_engine
from sqlalchemy import text
```

```python
# Conectar a una base de datos
engine = get_engine("basedeprueba")
# Mostrar el nombre de la base de datos conectada
print(engine.url.database)
```

```python
query = "SELECT * FROM empleados"
df = pd.read_sql(query, engine)
display(df)
```

```python
instruccion = text(
    """
    WITH RECURSIVE jerarquia AS (
    -- 1️⃣ Nivel base: el jefe inicial
    SELECT 
        id,
        nombre,
        id_jefe,
        1 AS nivel
    FROM empleados
    WHERE id = 2   -- Luis

    UNION ALL

    -- 2️⃣ Parte recursiva: buscar subordinados del empleado actual
    SELECT 
        e.id,
        e.nombre,
        e.id_jefe,
        j.nivel + 1
    FROM empleados e
    INNER JOIN jerarquia j
        ON e.id_jefe = j.id
    )
    SELECT *
    FROM jerarquia
    ORDER BY nivel;
    """
)

df = pd.read_sql(instruccion, engine)
display(df.style.hide())
# print(df.to_string(index=False))
```
